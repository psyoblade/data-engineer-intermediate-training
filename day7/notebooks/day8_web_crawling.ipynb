```python
genre_dic = {}

def makeUrl(date, kinds):
  if kinds == 'playstore': 
    url = 'https://www.mobileindex.com/app/get_rank_all?rt=r&mk=2&c=kr&t=app&rs=100&d=' + date
  else:
    url = 'https://www.mobileindex.com/app/get_rank_all?rt=r&mk=1&c=kr&t=app&rs=100&d=' + date
  return url
```

```python
url = makeUrl("2020-08-01", "playstore")
print(url)

import requests
r = requests.get(url)
print(r.text[:100])
```

```python
from bs4 import BeautifulSoup
html_doc = r.text
soup = BeautifulSoup(html_doc, 'html.parser')
# print(soup.prettify()) # ìš©ëŸ‰ì´ ì»¤ì„œ ë§‰ì•„ ë‘ì—ˆìŠµë‹ˆë‹¤.
```

```python
for span in soup.find_all('span')[:5]:
    if span.get("class")[0] == "appname":
        print(span)
```

```python
tables = soup.find_all('div', attrs={'class':'item-info'})

for table in tables:
    head = table.find('span', attrs={'class':'appname'}).text
    body = table.find('span', attrs={'class':'publisher'}).text
    print("%s => %s\n" % (head, body))
```
