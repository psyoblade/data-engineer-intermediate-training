### 1.2.1 Daily Confirmed Patients (일별 확진자수)

먼저, 일별 확진자수를 알아보기 위해 Time.csv 파일을 가져와 전처리 작업을 했습니다.

```python
import pandas as pd
from plotnine import *
from mizani.breaks import date_breaks
from mizani.formatters import date_format
from datetime import datetime
from math import sqrt
import numpy as np
import warnings
warnings.filterwarnings('ignore')
import matplotlib.font_manager as fm
font_location = 'C:/WINDOWS\/Fonts\/KBIZ한마음고딕 R.ttf'
font_name = fm.FontProperties(fname = font_location).get_name()
from adjustText import adjust_text
```

```python
time = pd.read_csv('COVID_Time.csv')
time['date'] = [datetime.strptime(x, '%Y-%m-%d') for x in time['date']] #change to datetime object for better visualization
daily =[0] + [x -y for x,y in zip(time['confirmed'][1:],time['confirmed'])] #daily confirmed patients
time['daily'] = daily
```

```python
(ggplot(data=time,mapping=aes(x='date',y='daily',group=1))
+geom_line(color='#E9A3B2',size=1)
+ggtitle("Daily Confirmed")
+scale_x_date(labels=date_format('%m/%d'),breaks=date_breaks('2 weeks'))
+theme(figure_size=(20,15),text=element_text(family=font_name))
+geom_text(aes(x='date',label='daily'),va='top',nudge_y=1,size =7)
+annotate(geom='text',x=datetime.strptime('2020-06-01','%Y-%m-%d'),y=100,label='Weak Social Distancing-코로나 안정기',va='bottom',size=20)
+geom_rect(mapping=aes(xmin=datetime.strptime('2020-04-20','%Y-%m-%d'),
              xmax = datetime.strptime('2020-07-01','%Y-%m-%d'),
              ymin=-5,ymax=100),fill ='#E3E4EA',alpha=0.01)
+annotate(geom='text',x=datetime.strptime('2020-03-31','%Y-%m-%d'),y=200,label='코로나 확산기!!!',va='bottom',size=20)
)
```

7월 이전의 국내 코로나 일일 확진자수의 추이는 위의 그래프와 같습니다. <br><br>

# 2. Risk Index (타격 지수)

앞서 타격 지수를 '(1) 코로나 확산기 동안 (2) 해당 업종의 소비가 얼마큼 많이 증가/감소했는지를 나타내는 지수'라고 설명했습니다. <br><br>
조금 자세히 설명드리자면, <br>
__(1) 코로나 확산기 동안__ <br>
1~3월 까지의 소비 변화를 분석했습니다.<br>
__(2) 해당 업종의 소비 변화__ <br>
1에서 2월 그리고 2에서 3월의 소비 비율 변화의 기하 평균을 산출했습니다 <br>

사용한 식은 다음과 같습니다.(R1은 1-2월의 소비 변화율, R2는 2-3월 소비 변화율) <br><br>
<script type="text/javascript" async src="https://cdnjs.cloudflare.com/ajax/libs/mathjax/2.7.5/MathJax.js?config=TeX-AMS_HTML" ></script>

<a href="https://www.codecogs.com/eqnedit.php?latex=Risk&space;Index&space;=&space;\sqrt{(R_1&plus;1)*(R_2&plus;1)}-1" target="_blank"><img src="https://latex.codecogs.com/png.latex?Risk&space;Index&space;=&space;\sqrt{(R_1&plus;1)*(R_2&plus;1)}-1" title="Risk Index = \sqrt{(R_1+1)*(R_2+1)}-1" /></a> <br><br>

기하 평균을 사용한 이유는, 소비량의 변화를 '수익'의 관점에서 보았기 때문입니다.<br> 
투자할 때 수익률의 평균은 보통 기하 평균을 사용합니다. 기하 평균의 수치가 다소 직관적이지 않을 수는 있지만, 복리를 가정하기 때문에 수익 변화의 연속성을 잡아낼 수 있고, '손실'을 잘 반영한다는 점에서 더 효과적이고 일반적인 평가 방법입니다.** <br><br><br>
** Levy H., <i>Stochastic Dominance: Investment Decision Making under Uncertainty</i>,  356-365, 2015<br>
*** 소비금액 적자(취소)를 기록한 항공 업종의 경우, ValueError를 피하기 위해 음수 값은 0에 아주 가까운 1e-6 값으로 변환해주었습니다.

```python
sales_tmp = sales.copy()
sales_tmp.loc[sales.salamt<0,'salamt'] = 1e-6
```

```python
(ggplot(sales_tmp[sales_tmp.mrhst_induty_cl_nm.isin(['항공','홈쇼핑'])])
+geom_line(aes(x='receipt_dttm',y= 'salamt',group='mrhst_induty_cl_nm',color='mrhst_induty_cl_nm'),size=1)
+theme(text=element_text(family=font_name))
+annotate(geom='text',x=2.5,y=20000000000,label='하락ㅠㅠ',va='top') 
+annotate(geom='text',x=2.5,y=40000000000,label='성장!!!',va='bottom')
+annotate(geom='text',x=2,y=5000000000,label='<----코로나 확산기---->',va='top') 
+ggtitle('Risk 지수 설명')
+geom_rect(mapping=aes(xmin=1,xmax =3, ymin=0,ymax=85000000000),fill ='#E3E4EA',alpha=0.01)
+scale_colour_hue(l=0.8)
)
```

> 쉽게 설명하기 위해 위의 그래프를 보시면, 1월 ~3월 '코로나 확산기' 동안 홈쇼핑의 소비는 성장하는데 반해, 항공의 소비는 급격하게 줄었습니다. 이러한 소비의 변화를 잡아내기 위한 지수가 바로 'Risk Index'입니다. <br>
Risk Index는 0보다 작으면 소비 감소, 0 보다 크면 소비 증가를 나타내며, 절대값의 크기가 클 수록 정도가 심하기 때문에 소비 변화를 손쉽게 표현할 수 있습니다.

```python
#Risk 지수 만들기
cat = sales[sales.receipt_dttm==2]['mrhst_induty_cl_nm']
risk2 = [x/y for x,y in zip(sales_tmp[sales_tmp.receipt_dttm==2]['salamt'],sales_tmp[sales_tmp.receipt_dttm==1]['salamt'])]
risk3 = [x/y for x,y in zip(sales_tmp[sales_tmp.receipt_dttm==3]['salamt'],sales_tmp[sales_tmp.receipt_dttm==2]['salamt'])]
risk_data = pd.DataFrame({'cat':cat,'risk_2':risk2,'risk_3':risk3})
risk_data['risk_index'] = [sqrt(x*y)-1 for x,y in zip(risk_data.risk_2,risk_data.risk_3)]
```

# 3. Resilience Index (탄력성 지수)

앞서 탄력성 지수를 '(1)코로나 안정기 동안 (2)해당 업종의 소비가 "얼마큼 빨리" 증가/감소했는지를 나타내는 지수'라고 설명했습니다. 더 자세히는 매출이 증가한 업종은 증가분이 지속 가능한지를 파악하고, 매출이 감소한 업종은 얼마큼 빨리 회복할 수 있을 지를 파악하고자 했습니다. 증가한 업종이나 타격을 입은 업종이 코로나 이전으로 돌아가는 것을 '탄력성'에 빗대어 이름을 Resilience Index라고 지었습니다.<br><br>
__(1) 코로나 안정기 동안__ <br>
3~5월의 데이터를 살펴보았습니다.<br>
__(2) 소비가 얼만큼 빨리__<br>
3월에서 4월, 그리고 4월에서 5월의 소비 금액 변화의 산술 평균을 살펴보았습니다. <br>
<br><br>
식을 살펴봅시다. (R1과 R2는 각각 3-4월 4-5월의 소비 변화율)<br><br>
<a href="https://www.codecogs.com/eqnedit.php?latex=Resilience&space;Index&space;=&space;\frac{(R_1&space;&plus;&space;R_2)}{2}" target="_blank"><img src="https://latex.codecogs.com/png.latex?Resilience&space;Index&space;=&space;\frac{(R_1&space;&plus;&space;R_2)}{2}" title="Resilience Index = \frac{(R_1 + R_2)}{2}" /></a><br>
<br><br>
 이번에는 산술 평균을 사용했는데요, 변화의 속도, 즉, '기울기'에 중점을 주었기 때문입니다. 앞서 말씀 드렸 듯이, 기하 평균은 자칫 비직관적일 수 있고, 손실에 민감합니다. 또한 여기서는 변화의 연속성보다는 개별 변화의 추이에 중점을 두고 싶었기에 산술 평균을 사용했습니다.

이 아이디어는 아래의 KBS1 '코로나19 통합 뉴스룸' 보도를 보면서 생각해냈습니다.<br><br>
<iframe src="https://drive.google.com/file/d/1k3kLW8pMmMvcMZUdGJKwI67jfMCsLjTh/preview" width="640" height="480"></iframe> <br>
위 사진을 보시면 기울기가 크면 경제가 회복되는 것에 긍정적인 기대를, 기울기가 작으면 경제 회복이 어렵다는 전망을 직관적으로 생각 할 수 있습니다. <br>
반대로 기울기가 크면 기울기가 작은 산업보다 성장이 조금 더 지속 될 것임을 예측할 수 있습니다. <br>
저의 탄력성 지수는 이런 기울기를 표현한 것입니다. 

```python
(ggplot(data=sales[sales.mrhst_induty_cl_nm.isin(['숙박업','면세점','홈쇼핑'])])
+geom_line(aes(x='receipt_dttm',y='salamt',group='mrhst_induty_cl_nm',color='mrhst_induty_cl_nm'),size=1)
+theme(text=element_text(family=font_name))
+annotate(geom='text',x=4.5,y=40000000000,label='코로나 이전으로..?',va='bottom')
+annotate(geom='text',x=4.5,y=20000000000,label='원점으로 복귀중!',va='top') 
+annotate(geom='text',x=4.5,y=2000000000,label='ㅠㅠ...',va='bottom')  
+ggtitle('Resilience 지수 설명')
+geom_rect(mapping=aes(xmin=3,xmax =5, ymin=0,ymax=50000000000),fill ='#E3E4EA',alpha=0.01)
+annotate(geom='text',x=4,y=50000000000,label='<-----코로나 안정기----->',ha='center',va='bottom')
+scale_color_hue(l=0.7)
)
```

> 쉽게 알아보기 위해 3-5월 동안 위 그래프의 세 업종의 변화를 봅시다. 
- 홈쇼핑의 경우, 코로나 이후 매출이 상승했지만 급격하게 다시 하락하고 있습니다. 포스트 코로나 시대에 지속가능한 매출 증대는 어렵다고 판단할 수 있겠습니다.
- 숙박업의 경우, 코로나로 인해 피해를 입었지만 급격하게 다시 회복을 하고 있습니다.
- 반면 면세점의 경우, 코로나로 큰 타격을 입었지만 회복 전망이 좋지 않습니다.
</ul><br>
이렇듯 기울기가 양수인지 음수인지, 그리고 큰지 작은지에 따라 포스트 코로나 시대에 업종의 소비 변화 짐작이 가능합니다.

```python
#Resilience Index 만들기
res4 = [(x-y)/y for x,y in zip(sales[sales.receipt_dttm==4]['salamt'],sales[sales.receipt_dttm==3]['salamt'])]
res5 = [(x-y)/y for x,y in zip(sales[sales.receipt_dttm==5]['salamt'],sales[sales.receipt_dttm==4]['salamt'])]
res_data = pd.DataFrame({'cat':cat,'res4':res4,'res5':res5})
res_data['res_index'] = (res_data.res5 + res_data.res4)/2
res_data.loc[res_data.cat=='항공','res_index'] =1
res_data['res_index_abs'] = abs((res_data.res5 + res_data.res4)/2)
res_data.loc[res_data.cat=='항공','res_index_abs'] =1
```

# 4. Risk -Resilience Matrix (타격-탄력성 분석)

이제 타격과 탄력성, 두 가지 지표로 산업을 묶어 봅시다.
두 가지 지표를 함께 보기 위해 2차원의 좌표 평면에 수치를 나타냈습니다.

```python
total = risk_data.join(res_data.set_index('cat'),on='cat')
volume = sales.groupby(['mrhst_induty_cl_nm']).mean()['salamt'].reset_index()
volume.columns=['cat','salamt']

total = total.join(volume.set_index('cat'),on='cat')
```

```python
cond = [
    (total.risk_index >= 0) & (total.res_index>= 0.07),
    (total.risk_index >= 0)&(total.res_index<= 0.07),
    (total.risk_index < 0)&(total.res_index>= 0.07),
    (total.risk_index < 0)&(total.res_index< 0.07)
]
values = ['SilverLining','Immune','Resilient','DoDo']
total['type'] = np.select(cond,values)
```

```python
(ggplot(data=total)+geom_point(aes(x='risk_index',y='res_index',fill='type'),size=4)
+theme(figure_size=(8,8),text=element_text(family=font_name))
+geom_hline(aes(yintercept=0.07),color='#586BDA')
+geom_vline(aes(xintercept=0),color='#DA6E58')
+ggtitle('Risk-Resilience Matrix')
+scale_fill_hue(l=0.7)
)
```

> 모든 업종을 Risk-Resilience Matrix에 표현해봤습니다.(이하 R-R Matrix) 분석의 편의상 탄력성 지수는 0.07을 기준으로 분류했습니다. 업종을 좌표평면 상의 위치를 기준으로 아래와 같이 총 4가지로 분류를 해보았는데요, <br><br>
>1. Silver Lining : 성장 + 지속
>2. Immune : 성장 + 하락세
>3. Resilient: 하락 + 빠른 회복
>4. DoDo : 하락 + 느린 회복</ol> <br>
<br>
